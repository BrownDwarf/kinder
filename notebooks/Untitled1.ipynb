From Grankin et al. 2008 Figure 5, we have that the amplitude of LkCa4 is as high as 0.79 magnitudes peak-to-valley, or as low as 0.35.  It varies season-to-season!  The average peak-to-valley magnitude difference is 0.512.  So converting that magnitude to a flux:

```python
f_ratio = 10.0**(0.512/2.5)
```

```python
f_ratio
```
